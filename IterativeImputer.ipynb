# Mice 補值

```python
import pandas as pd 
import numpy as np
```

```python
train_df=pd.read_csv('data/train.csv')
```

```python
test_df=pd.read_csv('data/test.csv')
```

```python
all_data=pd.concat([train_df,test_df])
```

```python
all_data_X=all_data.iloc[:,:-1]
```

```python
all_data_y=all_data.iloc[:,-1]
```

```python
# select the numeric columns
df_num = all_data_X.select_dtypes(include=[np.number])
# select non-numeric columns
df_non_num = all_data_X.select_dtypes(exclude=[np.number])
```

```python
df_num
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import xgboost

data=df_num.copy()

#setting up the imputer

imp = IterativeImputer(
    estimator=xgboost.XGBRegressor(
        n_estimators=5,
        random_state=1,
        #tree_method='gpu_hist',
    ),
    missing_values=np.nan,
    max_iter=30,
    initial_strategy='mean',
    imputation_order='ascending',
    verbose=2,
    random_state=1
)

data[:] = imp.fit_transform(data)
data.shape
```

```python
df_num_imp=data
```

```python
all_data_X_imp=pd.concat([df_num_imp,df_non_num,all_data_y],axis=1)
```

```python
all_data_X_imp
```

```python
all_train_new=all_data_X_imp[~all_data_X_imp['failure'].isnull()]
all_train_new.to_csv('data/train_imp.csv',index=False)
```

```python
all_test_new=all_data_X_imp[all_data_X_imp['failure'].isnull()]
all_test_new.to_csv('data/test_imp.csv',index=False)
```
