```python
import pandas as pd 
import numpy as np
```

```python
train_df=pd.read_csv('data/train_imp.csv')
```

```python
test_df=pd.read_csv('data/test_imp.csv')
```

```python
all_df=pd.concat([train_df,test_df])
```

```python
df_non_num = all_df.select_dtypes(exclude=[np.number])
df_non_num
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
enc.fit(df_non_num)
df_non_num_new=enc.transform(df_non_num)
```

```python
df_non_num_new=pd.DataFrame(df_non_num_new,columns=df_non_num.columns)
df_non_num_new
```

```python
all_df[df_non_num.columns]=df_non_num_new
all_df
```

```python
all_df=all_df.iloc[:,1:]
```

```python
train_df=all_df[~all_df['failure'].isnull()]
train_df
```

```python
test_df=all_df[all_df['failure'].isnull()]
test_df
```

```python
train_X=train_df.iloc[:,:-1]
train_y=train_df.iloc[:,-1]
```

```python
from xgboost import XGBClassifier

# 建立 XGBClassifier 模型
xgboostModel = XGBClassifier(
        eval_metric="logloss",
        max_depth=11,
        learning_rate=0.0642,
        n_estimators=800,
        min_child_weight=7,
        grow_policy="depthwise",
        gamma=1.2840e-06)
# 使用訓練資料訓練模型
xgboostModel.fit(train_X, train_y)
# 使用訓練資料預測分類
```

```python
predicted = xgboostModel.predict(train_X)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(train_y, predicted)
accuracy
```
